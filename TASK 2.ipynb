STEP 1: Import all the Libraries

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.patches as mpatches
import sklearn.metrics as sm
from mpl_toolkits.mplot3d import Axes3D
from scipy.cluster.hierarchy import linkage,dendrogram
from sklearn.cluster import DBSCAN 
from sklearn.decomposition import PCA 
```

```python
iris = datasets.load_iris()
print(iris.data)
```

Step 1 - Loading the dataset

```python
print(iris.target_names)
```

```python
print(iris.target)
```

iris.target provides the integer labels for the samples, which are used in classification tasks to train and evaluate models.

```python
x = iris.data
y = iris.target
```

x is used as the input to machine learning models for training or prediction.
y is the target variable that models aim to predict. It is used during training to help the model learn the mapping from input features to the target labels.

 Step 2 - Visualizing the input data and its Hierarchy

```python
fig = plt.figure(1, figsize=(7,5))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(x[:, 3], x[:, 0], x[:, 2], edgecolor="k", s=50)
ax.set_xlabel("Petal width")
ax.set_ylabel("Sepal length")
ax.set_zlabel("Petal length")
plt.title("Iris Clustering K Means=3", fontsize=14)
plt.show()


#Hierachy Clustering 
hier=linkage(x,"ward")
max_d=7.08
plt.figure(figsize=(15,8))
plt.title('Iris Hierarchical Clustering Dendrogram')
plt.xlabel('Species')
plt.ylabel('distance')
dendrogram(
    hier,
    truncate_mode='lastp',  
    p=50,                  
    leaf_rotation=90.,      
    leaf_font_size=8.,     
)
plt.axhline(y=max_d, c='k')
plt.show()
```

The first section of the code visualizes the Iris dataset in a 3D space to give an idea of how the data is distributed across the three chosen dimensions.

The second section visualizes the hierarchical clustering of the Iris dataset with a dendrogram. The dendrogram shows how clusters are formed and merged step by step. The horizontal line (plt.axhline) indicates a threshold for cutting the dendrogram to form flat clusters.

STEP 3 - Data Preprocessing

```python
x = pd.DataFrame(iris.data, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
y = pd.DataFrame(iris.target, columns=['Target'])
x.head()
```

```python
y.head()
```

 STEP 4 - Model Training 

```python
kmeanS_model = KMeans(n_clusters=3)
kmeanS_model.fit(x)

print(kmeanS_model.labels_)
```

```python
print(kmeanS_model.cluster_centers_)
```

 STEP 5 - Visualizing the Model Cluster 

```python
plt.figure(figsize=(14,6))

colors = np.array(['blue', 'yellow', 'red'])

predictedY = np.choose(kmeanS_model.labels_, [1, 0, 2]).astype(np.int64)

plt.subplot(1, 2, 1)
plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[y['Target']])
plt.title('Before classification')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[predictedY])
plt.title("After Model's classification")
plt.legend()
```

 STEP 6 - Calculating the Accuracy 

```python
sm.accuracy_score(predictedY, y['Target'])
```

The accuracy score of 0.44 indicates that the model's performance is poor, and there's significant room for improvement in predicting the correct classes of the Iris dataset.

Confusion Matrix

```python
sm.confusion_matrix(predictedY, y['Target'])
```
